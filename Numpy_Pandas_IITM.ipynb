```python
#PCA
def PCA(X):
  d,n = X.shape
  X -= X.mean(axis=1,keepdims=True)
  cov_ =X @ X.T/n
  eigenvalues,eigenvectors = np.linalg.eigh(cov_)
  var =np.flip(eigenvalues)
  pcs =np.flip(eigenvectors,axis=1)
  return var,pcs

def plot(X,pcs):
  d,n = X.shape
  plt.scatter(X[0,:], X[1,:],alpha=0.5)

  plt.axhline(color='black',linestyle ='--',linewidth = 0.8)
  plt.axvline(color='black',linestyle ='--',linewidth = 0.8)
  plt.axis ='equal'


  plt.xlim(np.min(X[0,:]),np.max(X[0,:]))
  plt.ylim(np.min(X[1,:]),np.max(X[1,:]))

  #pc1
  m_1 = pcs[:,0][1]/pcs[:,0][0]
  m_2 = pcs[:,1][1]/pcs[:,1][0]

  x = np.linspace(-2,2)
  plt.plot(x,m_1*x,color='green',label='PC1')
  plt.plot(x,m_2*x,color='red',label='PC2')
  plt.legend()
  plt.title('PCA')
  plt.xlabel('X-axis')
  plt.ylabel('Y-axis')

```

```python
mu = np.array([2,5])
cov = np.array([[1,0.9],[0.8,1]])
X = rng.multivariate_normal(mu, cov, size=100).T
var,pcs = PCA(X)
plot(X,pcs)
```

Sample covariance vs Population covariance

```python
def generate(mu,cov,n):
  X = rng.multivariate_normal(mu,cov,size =n).T
  return X

def sample_covariance(X):
  d,n = X.shape
  mu= X.mean(axis=1).reshape(d,1)
  X -=mu
  return X @ X.T/n-1

def population_covariance(X):
  d,n = X.shape
  mu= X.mean(axis=1).reshape(d,1)
  X -=mu
  return X @ X.T/n
```

```python
def generate(mu, cov, n):
    X = rng.multivariate_normal(mu, cov, size = n).T
    return X

def sample_covariance(X):
    d, n = X.shape
    mu = X.mean(axis = 1).reshape(d, 1)
    X -= mu
    return X @ X.T / n
```

```python
mu = np.array([1, 2])
# population covariance matrix
cov = np.array([
    [1, 0],
    [0, 5]
])

# Run experiment
T = 10 # number of runs for each n
norms = [ ]
n_vals = np.arange(1000, 100_000, 1000)
for n in n_vals:
    norm = 0
    for t in range(T):
        X = generate(mu, cov, n)
        C = sample_covariance(X)
        norm += np.linalg.norm(C - cov)
    norms.append(norm / T)

# Plot
plt.plot(n_vals, norms)
plt.title('Norm vs Dataset size')
plt.xlabel('Dataset size')
plt.ylabel('Norm');
```

Unconstrained Optimisation

SciPyâ€™s optimization routines are in the form of minimizers, so we will negate the objective function and minimze it.

```python
from scipy import optimize
def f(x):
  return -np.exp(-((x[0]+2)**2 + (x[1]-1)**2)/6)
res = optimize.minimize(f,x0=[0,0])
res.x
```

Constrained Optimisation

```python
from scipy import optimize
from scipy.optimize import LinearConstraint

# objective
def f(x):
    return x[0] ** 2 + x[1] ** 2 - 1

# optimize
res = optimize.minimize(f, np.zeros(2),
                        constraints = LinearConstraint(
                        A = np.array([1, 1]),
                        lb = 1))
res.x
```

```python
# contours of the objective
x_ = np.linspace(-4, 4, 100)
y_ = np.linspace(-4, 4, 100)
x, y = np.meshgrid(x_, y_)
cs = plt.contour(x, y, -f([x, y]), levels = [-1, -0.5, 0, 0.5, 1])
plt.clabel(cs)
# constraints
plt.plot(x_, 1 - x_, linestyle = '--', color = 'black')
# optimum
plt.scatter(0.5, 0.5)
# adjust plot
plt.axis='equal'
plt.xlim(-2, 2)
plt.ylim(-2, 2)
```
